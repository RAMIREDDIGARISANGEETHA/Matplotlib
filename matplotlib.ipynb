```python
#Bar Plot
```

```python
langs=['C','JAVA','PYTHON','HTML']
students=[43,12,50,31]
plt.bar(langs,students)
plt.xlabel('langs')
plt.ylabel('students')
plt.show()
```

```python
plt.bar([0.25,61.25,52.25,34.25,41.25],[50,40,70,80,20],
label="BMW",width=45)
plt.bar([5.75,8.75,2.75,53.75,41.75],[80,20,20,50,60],
label="Audi",color='g',width=6)
plt.title('Information')
plt.xlabel('Days')
plt.ylabel('Distance(kms)')
plt.legend()
plt.show()
```

```python
year=[2010,2011,2012,2013,2014]
width=0.25
indices=np.arange(len(year))
population_china=[1.359,1.397,1.404,1.414,1.555]
population_india=[2.359,2.397,2.404,2.414,2.555]
plt.bar(indices-width,population_india,width=width,label="india",color="green")
plt.bar(indices-width,population_china,width=width,label="china",color="red")
plt.title("Population Comparision between india and china")
plt.xlabel('Year')
plt.ylabel('Population in Billions')
plt.legend()
plt.xticks(ticks=indices,labels=year)
plt.tight_layout()
plt.grid(True)
plt.show()
           
```

```python
from matplotlib import pyplot as plt
import numpy as np
slices=[1234,6789,5543]
labels=['China','India','Pakisthan',]
explode=[0,0,0]
colors=['green','yellow','blue','white']
plt.pie(slices,labels=labels,shadow=True,startangle=90,
        wedgeprops={'edgecolor':'black','linewidth': 4},colors=colors)
plt.title("Population Pie Chart")
plt.grid(True)
plt.show()
```

```python
x=[1,1.3,3,2,4]
y=[5,6,3,2,4]
x1=[4,5,3,2,2]
y1=[7,2,3,2,5]
plt.scatter(x,y,label="high income low saving",color='r')
plt.scatter(x1,y1,label="low income high saving",color='b')
plt.xlabel('saving*100')
plt.ylabel('income*1000')
plt.title('ScatterPlot')
plt.legend()
plt.show()
```

```python
#Violin Plot
```

```python
collectn_1=np.random.normal(100,10,200)
collectn_2=np.random.normal(80,30,200)
collectn_3=np.random.normal(90,20,200)
collectn_4=np.random.normal(70,25,200)

#combining these different collection into list
data_to_plot=[collectn_1,collectn_2,collectn_3,collectn_4]
#create a figure instance
fig=plt.figure()
#create an axes instance
ax=fig.add_axes([0,0,1,1])
#create the boxplot
bp=ax.violinplot(data_to_plot)
plt.show()
```
